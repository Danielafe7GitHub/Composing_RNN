```python
#!/usr/bin/python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
from nltk.tokenize import RegexpTokenizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

Cargado NRC Lexicon de Emociones

```python
n = open("nrc.txt", 'r+')
linea = n.readline()
pal_senti_emoci = []
nrc_lexicon = []
```

Transformando Lexicon a vectores

```python
#No olvides :Añadir un enter al ultimo elemento de nrc.txt
while linea != "\n":
    tokens = nltk.word_tokenize(linea)
    for token in tokens:
        pal_senti_emoci.append(token)
    nrc_lexicon.append(pal_senti_emoci)
    pal_senti_emoci = []
    linea = n.readline()
n.close()
```

Definiendo 7 emociones básicas + 2 sentimientos

```python
positive = 0 
negative = 0
anger = 0 
fear = 0 
surprise = 0 
sadness = 0 
joy = 0  
disgust = 0
```

Funcion de conteo de emociones

```python
def _nrc(pal_novela):
    #print ("Pal analizada:  "+pal_novela)
    global positive 
    global negative 
    global anger  
    global fear  
    global anticipation 
    global trust 
    global surprise 
    global sadness 
    global joy  
    global disgust
    global num_pal 
    global num_pal_emo 
    global emociones
 
    for emoWord in nrc_lexicon:
        if emoWord[0] == pal_novela:
            num_pal_emo += 1
            if emoWord[1] == "1":
                positive += 1
            if emoWord[2] == "1":
                negative += 1
            if emoWord[3] == "1":
                #print("anger"," con ", pal_novela)
                anger += 1
                emociones[0] += 1
            #if emoWord[4] == "1":
                #anticipation += 1
            if emoWord[5] == "1":
                #print("disgust"," con ", pal_novela)
                disgust += 1
                emociones[1] += 1
            if emoWord[6] == "1":
                #print("fear"," con ", pal_novela)
                fear += 1
                emociones[2] += 1
            if emoWord[7] == "1":
                #print("joy"," con ", pal_novela)
                joy += 1
                emociones[3] += 1
            if emoWord[8] == "1":
                #print("sadness"," con ", pal_novela)
                sadness += 1
                emociones[4] += 1
            if emoWord[9] == "1":
                #print("surprise"," con ", pal_novela)
                surprise += 1
                emociones[5] += 1
            #if emoWord[10] == "1":
                #trust += 1
```

Etiquetamos las oraciones

```python
def label_sentence(maximo,indice):
    global l_anger
    global l_fear
    global l_surprise
    global l_sadness 
    global l_joy  
    global l_disgust
    
    if maximo == 0:
        print("None")
    elif indice == 0:
        print("Label sentence is anger")
        l_anger += 1
    elif indice == 1:
        print("Label sentence is disgust")
        l_disgust += 1
    elif indice == 2:
        print("Label sentence is fear")
        l_fear += 1
    elif indice == 3:
        print("Label sentence is joy")
        l_joy += 1
    elif indice == 4:
        print("Label sentence is sadness")
        l_sadness += 1
    elif indice == 5:
        print("Label sentence is surprise")
        l_surprise += 1
```

```python
def _max(lista):
    count = 0
    e_maximo = max(lista)
    indice = lista.index(max(lista))
    
    if e_maximo == 0:
        return 0,indice
    else:
        for e in lista:
            if e == e_maximo:
                count += 1
        if count == 1:
            return e_maximo,indice
        else:
            return 0,indice
```
